```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# BACA CITRA (grayscale)
img = cv2.imread('gambar_uts1.png', cv2.IMREAD_GRAYSCALE)  # ganti nama file sesuai datamu

plt.imshow(img, cmap='gray')
plt.title("Citra Asli (Grayscale)")
plt.axis('off')
plt.show()
```

```python
# TRANSFORMASI NEGATIF
negatif = 255 - img

plt.imshow(negatif, cmap='gray')
plt.title("Citra Negatif")
plt.axis('off')
plt.show()
```

```python
# TRANSFORMASI LOGARITMIK
img_float = img.astype(np.float32)

# hitung konstanta c
c = 255 / np.log(1 + np.max(img_float))

# lakukan transformasi log
log_image = c * np.log(1 + img_float)

# ubah kembali ke uint8
log_image = np.uint8(log_image)

plt.imshow(log_image, cmap='gray')
plt.title("Citra Hasil Transformasi Logaritmik")
plt.axis('off')
plt.show()
```

```python
# HISTOGRAM CITRA ASLI, NEGATIF, LOG
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.hist(img.ravel(), bins=256, color='black')
plt.title("Histogram Citra Asli")

plt.subplot(1, 3, 2)
plt.hist(negatif.ravel(), bins=256, color='black')
plt.title("Histogram Citra Negatif")

plt.subplot(1, 3, 3)
plt.hist(log_image.ravel(), bins=256, color='black')
plt.title("Histogram Citra Logaritmik")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Asli")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(negatif, cmap='gray')
plt.title("Negatif")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(log_image, cmap='gray')
plt.title("Logaritmik")
plt.axis('off')

plt.tight_layout()
plt.show()
```
