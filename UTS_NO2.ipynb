```python
import cv2
import numpy as np

# Baca gambar grayscale
img = cv2.imread('Uts_no2.jpg', 0)

# Tiga kernel sesuai soal
kernel_lpf = np.ones((3,3), np.float32) / 9
kernel_hpf = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
kernel_bpf = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

# Filtering
img_lpf = cv2.filter2D(img, -1, kernel_lpf)
img_hpf = cv2.filter2D(img, -1, kernel_hpf)
img_bpf = cv2.filter2D(img, -1, kernel_bpf)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img_lpf, cmap='gray')
plt.title("LPF (Blur)")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(img_hpf, cmap='gray')
plt.title("HPF (Edge Detection)")
plt.axis('off')

plt.subplot(2,2,4)
plt.imshow(img_bpf, cmap='gray')
plt.title("BPF (Sharpening)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
